```python
def sum_func(arg1, arg2, *args): # *args: optional extra arguments allowed and stored in list called args
    out_sum = arg1+arg2
    for i in args:
        out_sum += i
    return out_sum

sum_func(3, 4, 5, 6)
```

```python
def dict_func(arg1, arg2, **kwargs): #kwargs allow extra vars of the form a=b (a is key, b is value) and stores as a dictionary
    output_dict = {}
    output_dict[arg1] = arg2
    for i in kwargs.keys():
        output_dict[i] = kwargs[i]
    return output_dict

# dict_func("key1", 5)
dict_func("key1", 5, key2=6)
```

```python
def class_graph(student1: str, student2: str, *args, class_dict=class_dictionary):
    '''A function to generate a horizontal bar plot comparing student midterm scores.
    INPUTS: 
        student1 - name of a student from class_dictionary
        student2 - name of another student from class_dictionary
        *args (optional) - any number of students from class_dictionary
    PRODUCES: 
        A bar plot
    RETURNS: 
        NONE
    '''
    #class_dict is a pre-set or default argument of the func. must be defined after all other required, non-default arguments.
    len_plot = range(2 + len(args))
    score = [class_dict[student1], class_dict[student2]] + [class_dict[i] for i in args] 
    #list comprehension above. list addition below.
    names = [student1] + [student2] + [i for i in args]
    
    plt.barh(len_plot, score)
    plt.yticks(len_plot,names)
    plt.title('Comparison of student scores')
    plt.xlabel('Score')
    plt.ylabel('Student')
    plt.show()
    return 

class_graph('Sarah', 'Josiah')
class_graph('Sarah', 'Josiah', 'Malena')
```

```python
# even more generalized function, where user provides class_dictionary 

def compare_students(class_dict, student1: str, student2: str, *args):
    '''A function to generate a horizontal bar plot comparing student midterm scores.
    INPUTS: 
        class_dict - dictionary containing student names and exam scores of the form {'name' (str): score (float)}
        student1 - name of a student from class_dictionary
        student2 - name of another student from class_dictionary
        *args (optional) - any number of students from class_dictionary
    PRODUCES: 
        A bar plot
    RETURNS: 
        NONE
    '''
    len_plot = range(2 + len(args))
    score = [class_dict[student1], class_dict[student2]] + [class_dict[i] for i in args] 
    #list comprehension above. list addition below.
    names = [student1] + [student2] + [i for i in args]
    
    plt.barh(len_plot, score)
    plt.yticks(len_plot,names)
    plt.title('Comparison of student scores')
    plt.xlabel('Score')
    plt.ylabel('Student')
    plt.show()
    return 

compare_students(class_dictionary, 'Sarah', 'Josiah')
compare_students(class_dictionary, 'Sarah', 'Josiah', 'Malena')
```
